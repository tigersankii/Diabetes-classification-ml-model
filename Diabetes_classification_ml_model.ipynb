**IMPORTING THE PYTHON LIBRARIES**

```python
import numpy as np
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
```

UPLOADING THE DATASET

```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("akshaydattatraykhare/diabetes-dataset")

print("Path to dataset files:", path)
```

What-you-need-to-know-about-diabetes.jpg

LOADING THE DATASET

```python
df = pd.read_csv('/kaggle/input/diabetes-dataset/diabetes.csv')
df
```

SPLIT THEDTA INTP FEATURES (X) AND TARGET (Y)

```python
x = df[df.columns[:-1]]
x
```

```python
y = df["Outcome"]
y
```


PERFORM A TRAIN-TEST SPL

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=0)
```

```python
x_train
```

```python
y_train
```

TRAIN AND EVALUATE FOUR MODELS:



```python
model1 = KNeighborsClassifier()
model2 = GaussianNB()
model3 = LogisticRegression()
model4 = SVC()
```

```python
model1.fit(x_train,y_train)
model2.fit(x_train,y_train)
model3.fit(x_train,y_train)
model4.fit(x_train,y_train)
```

FOR EACH MODEL,PRINT CLASSIFICATION REPORT AND ACCURACY

```python
print(classification_report(y_test,model1.predict(x_test)))
```

```python
print(classification_report(y_test,model2.predict(x_test)))
```

```python
print(classification_report(y_test,model3.predict(x_test)))
```

```python
print(classification_report(y_test,model4.predict(x_test)))
```

Note: the best model is model3 (LogisticRegression)
